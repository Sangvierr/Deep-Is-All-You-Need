```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 이미지 로드
# Check if the image file exists
import os


# 이미지 로드
image = cv2.imread('./ex_image.jpg', cv2.IMREAD_GRAYSCALE)

# 수평선 필터 커널 정의
horizontal_line_filter = np.array([[1, 1, 1],
                                    [0, 0, 0],
                                    [-1, -1, -1]])

# 수평선 필터 커널 정의
filtered_image = cv2.filter2D(image, -1, horizontal_line_filter)

# 수직선 필터 커널 정의
filtered_image2 = cv2.filter2D(image, -1, horizontal_line_filter.T)

# 결과 시각화
plt.figure(figsize=(16, 9))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Filtered Image (Horizontal Line)')
plt.imshow(filtered_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Filtered Image (Vertical Line)')
plt.imshow(filtered_image2, cmap='gray')
plt.axis('off')

plt.show()
```
